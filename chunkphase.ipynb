```python
from nltk import pos_tag,RegexpParser
from nltk.tokenize import PunktSentenceTokenizer, word_tokenize,sent_tokenize
import string
import re
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from nltk.corpus import wordnet
from nltk.tag import pos_tag
from nltk import DependencyGraph
import pprint
nltk.download('punkt')
```

```python
puts=string.punctuation+'’' #python中的標點符號
def remove_punctuatuon(text):
    return text.translate(str.maketrans('','',puts))

stopwords_list=stopwords.words('english')
#print(stopwords_list)
def remove_stopwords(text):
    correct_words=[]
    for i in text.split():
        if i in stopwords_list:
            continue
        else:
            correct_words.append(i)
    return ' '.join(correct_words)

#定義斷句的函式
def word_tokenized(text):
    sentence_tokenized=sent_tokenize(text)
    word_tokenize=[]
    for i in sentence_tokenized:
        word_tokenize.append(i)
    return word_tokenize
with open(r'C:\downloded dataset\nlp article\chunk_phase.txt',encoding="utf-8") as f:
    text=f.read().lower()

text=re.sub('\“',' ',text)
text=re.sub('\—',' ',text)
text=re.sub('\”',' ',text)
#text=remove_stopwords(text)

sen_tokenized=word_tokenized(text)
tokens=[]
#斷詞
for i in sen_tokenized:
    t=word_tokenize(i)
    for  j in t:
        if j in puts:
            a=t.index(j)
            t.pop(a)
            continue
        else:
            tokens.append(j)
print(tokens)
```

```python
def form_speech_dict(pos):
    t={}
    for i in pos:
        t[i[0]]=i[1]
    return t
print(form_speech_dict(pos_tag(tokens)))
```

```python
np='NP:{<DT>?<JJ>*<NN.*>}'
vp="VP: {<DT>?<JJ>*<NN><VB.?><RB.?>?}"
np_parser=RegexpParser(np)
vp_parser=RegexpParser(vp)
result_np=np_parser.parse(pos_tag(tokens))
check = {}

for i in np_parser.parse(pos_tag(tokens)):
    i=list(i)
    l = (1,3)
    if (type(i[0])==type(l)):
        if (i[0] not in check.keys()):
            check[i[0]]=1
        else : check[i[0]]+=1
pprint.pprint(check)
#result_np.draw()
```
